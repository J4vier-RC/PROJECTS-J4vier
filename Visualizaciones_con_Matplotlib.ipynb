#Objetivo

## Analizar la distribución de los Registros de los establecimientos dados de alta en el Sistema de Información Empresarial Mexicano (SIEM) a nivel nacional con datos actualizados a fecha de agosto de 2024.

Descarga de datos

Los datos han sido cargados directamente a esta sesión de Colab. Se encuentran disponibles en el siguiente link:

Fuente: https://datos.gob.mx/busca/dataset/sistema-de-informacion-empresarial-mexicano-siem

```python
## IMPORTAR LIBRERIAS
import pandas as pd
```

```python
# Se crea un Dataframe con esta base de datos
datos = pd.read_csv('datos_emp.csv',  encoding='latin1')
# Se muestran las primeras 5 filas del DataFrame
datos.head()
```

```python
#Se borran algunas columnas que no serán utilizadas para este analisis
datos = datos.drop(columns=['Domicilio', 'Colonia', 'C.P.', 'Teléfono','E-mail'])
datos
```

```python
# Se eliminan filas con valores faltantes en cualquier columna
datos = datos.dropna()
datos
```

```python
# Eliminar filas donde la columna 'Giro' tenga el valor '.' para evitar un analisis
# erroneo, puesto que esta es información incompleta No es recomendable sustituir
# estos datos con el Giro más común puesto que se sesgarán los datos debido al gran
# número de registros con esta categoria erronea
datos = datos[datos['Giro'] != '.']
datos
```

```python
# Se obtienen los 7 Estados con más registros
top_estados = datos['Estado'].value_counts().nlargest(7)

# Se crea un DataFrame con los resultados
tabla_estados = pd.DataFrame({'Estado': top_estados.index, 'Cantidad': top_estados.values})

# Muestra la tabla
tabla_estados
```

```python
# Se cargan las librerias para comenzar a trabajar los graficos

import pandas as pd
import matplotlib.pyplot as plt

# Se crea una gráfica de barras que va a mostrar a los Estados con más empresas
# registradas en México

plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura
plt.bar(tabla_estados['Estado'], tabla_estados['Cantidad'], color='skyblue')

for i, v in enumerate( tabla_estados['Cantidad']  ):
    plt.text(i, v + 0.5, f'{v:.1f}', ha='center', va='bottom', fontsize=10)
plt.xlabel('Estado')
plt.ylabel('Número de empresas')
plt.title('Estados con más empresas en México, 2024')
plt.xticks(rotation=45, ha='right')  # Rota las etiquetas del eje x para mayor legibilidad
plt.tight_layout()  # Ajusta el diseño para evitar que las etiquetas se superpongan
plt.show()
```

```python
# Se obtienen los 7 giros empresariales con más registros de los 7 Estados seleccionados
top_giro = datos['Giro'].value_counts().nlargest(8)

# Se crea un DataFrame con los resultados
tabla_giro = pd.DataFrame({'Giro': top_giro.index, 'Cantidad': top_giro.values})
#tabla_giro = tabla_giro.drop(tabla_giro.index[0])

# Muestra la tabla
tabla_giro
```

```python
# Se crea una gráfica de barras
plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura
plt.bar(tabla_giro['Giro'], tabla_giro['Cantidad'], color='skyblue')

for i, v in enumerate( tabla_giro['Cantidad']  ):
    plt.text(i, v + 0.5, f'{v:.1f}', ha='center', va='bottom', fontsize=10)
plt.xlabel('Giro empresarial')
plt.ylabel('Número de empresas')
plt.title('Giros empresariales con más presencia en México, 2024')
plt.xticks(rotation=45, ha='right')  # Rota las etiquetas del eje x para mayor legibilidad
plt.tight_layout()  # Ajusta el diseño para evitar que las etiquetas se superpongan
plt.show()
```

Los resultados muestran que los giros empresariales más comunes en estos 7 Estados, son actividades economicas que no requieren recursos especializados para iniciar sus actividades. Es decir, se pueden categorizar en Micro, Pequeñas, y Medianas Empresas.

```python
# Se obtienen los 7 rangos de empleados con más registros
top_rango = datos['Rango de Empleados'] .value_counts().nlargest(7)

# Se crea un DataFrame con los resultados
tabla_rango = pd.DataFrame({'Rango Empleados': top_rango.index, 'Cantidad': top_rango.values})

# Muestra la tabla
tabla_rango
```

```python
# Se crea una gráfica circular
plt.figure(figsize=(8, 8))
#plt.pie(tabla_rango['Cantidad'], labels=None, autopct='%1.1f%%', startangle=90,  pctdistance= 0.85)  # Quita las leyendas por defecto
plt.pie(tabla_rango['Cantidad'], labels=tabla_rango['Rango Empleados'], autopct='%1.1f%%', startangle=90, pctdistance=0.85) #Aumenta la distancia


# Se crea una leyenda fuera del gráfico
plt.legend(tabla_rango['Rango Empleados'], title='Rango de Empleados', loc="center left", bbox_to_anchor=(1, 0.5))

plt.title('Porcentaje de Empresas por Rango de Empleados en México, 2024')
plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()
```

Los resultados anteriores sobre la abundancia de Micro Empresas y PyMEs se confirman al observar que más del 95% de estas Empresas registradas tienen por mucho 50 trabajadores en activo en sus plantillas.

```python
# Se hace un Dataframe con los datos de CDMX, debido a que este es la Entidad con
# más empresas registradas, por lo que es interesante conocer la distribución de esas
# empresas a Alcaldia
cdmx = pd.DataFrame(datos[datos['Estado'] == 'Ciudad de México'])
cdmx
```

```python
# Se obtienen los 7 "municipios" de la CDMX con más registros
top_municipio = cdmx['Municipio'].value_counts().nlargest(7)

# Se crea un DataFrame con los resultados
tabla_municipio = pd.DataFrame({'Municipio': top_municipio.index, 'Cantidad': top_municipio.values})

# Muestra la tabla
tabla_municipio
```

```python
# Se crea una gráfica de barras
plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura
plt.bar(tabla_municipio['Municipio'], tabla_municipio['Cantidad'], color='skyblue')

for i, v in enumerate( tabla_municipio['Cantidad']  ):
    plt.text(i, v + 0.5, f'{v:.1f}', ha='center', va='bottom', fontsize=10)
plt.xlabel('Alcaldías')
plt.ylabel('Número de empresas')
plt.title('Alcaldías con más empresas en la CDMX, 2024')
plt.xticks(rotation=45, ha='right')  # Rota las etiquetas del eje x para mayor legibilidad
plt.tight_layout()  # Ajusta el diseño para evitar que las etiquetas se superpongan
plt.show()
```

Se hizo un grafico de la distribución de las 7 Alcadías con más empresas registradas en la CDMX, de las cuales las principales Alcaldías son: Cuauhhtemoc, Miguel Hidalgo y Benito Juarez.

En conclusión, los datos de las Empresas Registradas son en su mayoria de Micro, Pequeñas, y Medianas Empresas. En la mayoria de los casos, estas empresas no tienen más de 50 trabajadores.

Asimismo, aunque no se poseen esos datos, la tendencia del Comercio informal y de las Empresas que no han sido registradas a lo largo del País, deben ser similares.
